# etape 3 : fichier avec Python
## fichier texte:

```python
print("Creation d'un fichier texte (il sera ecrasé si il est deja crée)")
fichier = open("texte.txt","wt")
fichier.write("coucou\n")
fichier.write("Je suis un fichier texte\n")
fichier.close()


print("Lecture du fichier texte")
fichier = open("texte.txt","rt")
print(fichier.readlines())
fichier.close()

print("Ajout d'une ligne a la fin du fichier")
fichier = open("texte.txt","at")
fichier.write("Je suis une ligne en plus")
fichier.close()

print("Lecture du fichier texte")
fichier = open("texte.txt","rt")
print(fichier.readlines())
fichier.close()
```

```python
print(f'le code ASCII de A est {ord("A")}, soit {hex(ord("A"))} en hexa')
print(f"le code ASCII de B est {ord('B')}, soit {hex(ord('B'))} en hexa")
```

```python
chaineA = 'éèàù'
print(chaineA)
print(f'La longueur de chaineA est : {len(chaineA)}') # Il y a 4 caractères
print("")
chaineB = chaineA.encode()
print(chaineB) # En hexa : C3A9 (é), C3A8 (è), C3A0 (à), C3B9 (ù)
print(f'La longueur de chaineB est : {len(chaineB)}') # Il y a 8 octets
# Chaque caractère spécial est codé sur deux octets.
print("")
print(chaineB.decode())
print("")

chaineC = chaineA.encode("8859") # On convertit notre chaîne UTF-8 en ISO-8859 (ASCII Latin)
print(chaineC)
print(f'La longueur de chaineC est : {len(chaineC)}')
```
